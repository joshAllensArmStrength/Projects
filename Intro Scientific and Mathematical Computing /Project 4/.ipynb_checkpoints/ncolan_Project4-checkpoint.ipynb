```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina' # Used to sharpen tartan image
```

```python
def createTartan(colorSequence,colorPixelWidth,resolution):
    # First, adjust colorSequence & colorPixelWidth to fit in 500x500 image:
    
    # If the pattern of colors is less than the required resolution, keep repeating the pattern until
    # it is atleast the size of the resolution
    if sum(colorPixelWidth) < resolution:
        for width,seq in zip(colorPixelWidth,colorSequence):
            if sum(colorPixelWidth) < resolution:
                colorPixelWidth.append(width)
                colorSequence.append(seq)
                
    # Now that the color pattern is either greater than or equal to the required resolution, check to see
    # if the pattern is greater than the resolution. If it is, reduce the width of the last color pattern to 
    # correctly fit into the resolution
        if sum(colorPixelWidth) > resolution:
            colorPixelWidth[-1] = colorPixelWidth[-1] - (sum(colorPixelWidth) - resolution)
            
    # Create 2 arrays to hold the vertical and horizontal stripes of the tartan       
    vertical = np.empty([resolution,resolution,3])
    horizontal = np.empty([resolution,resolution,3])
    
    # Fill the 'vertical' array with columns of stripes
    j = 0
    for color,width in zip(colorSequence,colorPixelWidth):
        vertical[:,j:j+width] = color
        j = j+width
    
    # Fill the 'horizontal' array with rows of stripes
    j = 0
    for color,width in zip(colorSequence,colorPixelWidth):
        horizontal[j:j+width,:] = color
        j = j+width
    
    # Create array to represent the final tartan, it will be of the same shape as the vertical and horizontal
    # arrays
    tartan = np.empty_like(vertical)
    
    # Now fill the tartan with colors...to get checkerboard pattern and a 1 pixel vertical/1 pixel horizontal pattern,
    # we pick vertical when both i and j % 2 == 0 or == 1 
    
    # To get a 2 vertical -> 2 horizontal pattern pattern instead of 1 vertical/1 horizontal, use % 4 < 2 to fill 
    # vertical pattern.
    for i in range(tartan.shape[0]):
        for j in range(tartan.shape[1]):
            if (((i+1)+(tartan.shape[1]-j)) % 4 < 2):
                tartan[i,j] = vertical[i,j]
            else:
                tartan[i,j] = horizontal[i,j]
                
    #plt.imsave('nColanTartan.png',tartan)
    
    plt.figure(figsize=(10,8))
    plt.imshow(tartan)
    plt.show()
```

```python
W = np.array([224,224,224])/256
T = np.array([96,64,0])/256
LT = np.array([160,136,88])/256
N = np.array([136,136,136])/256

colorSeq = [N,T,N,T,W,LT,T,LT,T,LT,W,T,N,T,N]
colorWidth = [2,4,30,4,20,30,4,4,4,30,20,4,30,4,2]

createTartan(colorSeq,colorWidth,500)
```
