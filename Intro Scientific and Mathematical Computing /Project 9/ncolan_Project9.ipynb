# Introduction

In this project we will look at several different sets of data and analyze the occurences of each number $1$-$9$ as the first significant digit of a number, as well as the last significant digit of a number. The data that we will be looking at will include NYSE stock prices, country populations over different years, and elevations of airports all around the world. We will finish by looking at an edition of the New York Times, and analyze every number in that day's paper.

We will start by defining two methods; `fd()` that will determine the first significant digit, and `ld()` that will determine the last significant digit of a number. The number $0$ will be excluded from our analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import re
```

```python
def fd(number):
    num = 0
    sNumber = str(number).replace('-','').replace('.','')
    for n in sNumber:
        if n.isdigit():
            if int(n) != 0:
                num = int(n)
                return num
        else:
            continue
    return num
```

```python
def ld(number):
    num = 0
    sNumber = str(number).replace('-','').replace('.','')
    if(len(sNumber)==1):
        return number
    for n in range(len(sNumber)-1,-1,-1):
        if sNumber[n].isdigit():
            if int(sNumber[n]) != 0:
                num = int(sNumber[n])
                break
        else:
            continue
    return num
```

Now let's create a dictionarys to keep track of every first digit in the columns for the open, high, low, and close for each stock.

```python
nyseFirstDigitOpen_dict = {i:0 for i in range(10)}
nyseFirstDigitHigh_dict = {i:0 for i in range(10)}
nyseFirstDigitLow_dict = {i:0 for i in range(10)}
nyseFirstDigitClose_dict = {i:0 for i in range(10)}
```

With dictionary's created to keep track of the first digit for open, high, low, and close values of the listed stocks, we can now iteratively go through each row of our desired NYSE dataframe columns and count how many times each number appears as the first digit. Each column of the datafram is a `Series` object, and to access each value we need to use `Series.iteritems()`. The method `updateDictInfo` will go through the desired column of the dataframe and update the corresponding dictionary digit count.

```python
def updateDictInfo(dFrame,columnName=str,d=dict):
    for i in dFrame[columnName].iteritems():
        d[i[1]]+=1
    return d
```

With `updateDictInfo` we can now update all of our dictionaries to have the correct information.

Looking at the data for the first digit of the NYSE trading data, we can see that there is a clear trend in how often a number appears as the first digit. The number $1$ appears as the first digit at a much higher frequency than any other number, with a declining frequency as the integer increases. This trend in our data set agrees with `Benford's Law`, which is an observation that in many real-life sets of numerical data, the leading digit is likely to be small. 

In data sets that obey `Benford's Law`, the number $1$ appears as the leading significant digit about $30$% of the time, while the number $9$ appears as the leading significant digit less than $5$% of the time. We will now create a function that will display the exact % occurence of each digit named `rateOfOccurence`. We want to exclude all potential instances of $0$ returned by `fd()` and `ld()` from our % calculations, which would be the $0$th index from `list(dictionary.values())`, so we will exclude the first index of our list in this calculation.

```python
def rateOfOccurence(d=dict):
    sumValues = sum(list(d.values())[1:])
    for i in range(1,len(d)):
        print('{:}: {:.2f}%'.format(i,(d[i]/sumValues)*100))
```

Now lets check the NYSE opening stock price first digit data:

```python
rateOfOccurence(nyseFirstDigitOpen_dict)
```

We can see further evidence that our NYSE data obeys `Benford's Law`, as the number $1$ appears about $29.8$% of the time while the number $9$ appears about $3.9$% of the time. We can also see a general dropoff in frequency as the first digit increases from $1$-$9$.

`Benford's Law` further goes on to state that the distribution of the `n`-*th* digit, as `n` increases, rapidly approaches a uniform distribution. Looking at our bar chart for the last digits, we can see that this is true when looking at our NYSE data. Although it is not exactly uniform, there is a generally uniform distribution of integers $1$-$9$ that appear as the last digit. Analyzing the occurences of the last digit in the opening NYSE stock price data:

```python
rateOfOccurence(nyseLastDigitOpen_dict)
```

As we can see, there is a generally uniform distribution for each digit $1$-$9$ as the last digit in our opening stock prices data.

# New York Times Article

We will now look at a New York Times article from $2010$ and analyze the frequency of each number as the first and last digit.

```python
nytURL = 'https://jllottes.github.io/_downloads/7a14d4bc3565f1af04f4b5d90cfb6bfc/NYT_2010_06_01.txt'
```

Convert url text from `bytes` into a `string`:

```python
nytText = str(requests.get(nytURL).content)
```

Use regex to find all numbers in the article and store in `numbers`:

```python
pattern = r'\.*\d[\d,.-]*'
numbers = re.findall(pattern,nytText)
```

```python
nyt_fd = {i:0 for i in range(10)}
nyt_ld = {i:0 for i in range(10)}
```

Now let's write a function `updateDict_fromList()` that can update our dictionary's digit count when using a `list[str]` as the data structure:

```python
def updateDict_fromList(func,nums=list,d=dict):
    for n in nums:
        n = n.replace('-','').replace(',','').replace('.','')
        d[func(float(n))]+=1
    return d
```

Now we can update our first and last digit count using `updateDict_fromList()`:

```python
nyt_fd = updateDict_fromList(fd,numbers,nyt_fd)

nyt_ld = updateDict_fromList(ld,numbers,nyt_ld)
```

Get the `values()` to use as y-values of graph:

```python
nyt_fdValues = list(nyt_fd.values())

nyt_ldValues = list(nyt_ld.values())
```

Generate x-values:

```python
nyt_Index = ['1','2','3','4','5','6','7','8','9']
nyt_x = np.arange(len(nyt_Index))
```

Now with our data set we can first graph the occurence of each number as the first digit:

```python
plt.figure(figsize=(10,6),dpi=100)

plt.bar(nyt_x,nyt_fdValues[1:],width=0.2,label='Numbers')

plt.xticks(nyt_x,nyt_Index)

plt.xlabel('Digits')
plt.ylabel('Frequency')
plt.title('New York Times - Occurence of Every Number as First Digit')
plt.legend(prop={'size': 10})
plt.show()
plt.tight_layout()
```

We can see here visually that all of the articles from this edtion of the New York Times follow `Benford's Law` for the first significant digit - smaller numbers represent higher occurences of all significant first digits. Now let's look at the exact % occurences of each number as the first digit:

```python
rateOfOccurence(nyt_fd)
```

From here we can see that the data hits our benchmarks from `Benford's Law`, with the number $1$ representing about $33$% of all occurences of first digits and the number $9$ representing less than $5$% of all first digit occurences. Now let's look at occurences of last significant digits:

```python
nyt_Index = ['1','2','3','4','5','6','7','8','9']
nyt_x = np.arange(len(nyt_Index))
```

```python
plt.figure(figsize=(10,6),dpi=100)

plt.bar(nyt_x,nyt_ldValues[1:],width=0.2,label='Numbers')

plt.xticks(nyt_x,nyt_Index)

plt.xlabel('Digits')
plt.ylabel('Frequency')
plt.title('New York Times - Occurence of Every Number as Last Digit')
plt.legend(prop={'size': 10})
plt.show()
plt.tight_layout()
```

```python
rateOfOccurence(nyt_ld)
```

There is a slight tendency for the numbers $1$, $2$, and $5$ to occur more often than not as the last digit, which occur slightly more than any other numbers. But we can conclude that there is generally a uniform occurence of each number outisde of those three. 

This tendency towards $1$, $2$, and $5$ can be partially explained by the occurence of single digit numbers in the article. Let's find how many single digit numbers there are:

```python
count = 0
for n in numbers:
    if len(n)==1:
        count +=1
        
print('{:} / {:} -- {:.2f}%'.format(count,len(numbers),(count/len(numbers))*100))
```

As we can see, there were $139$ single-digit numbers in this data set. These single digit numbers are treated the same way by `ld()` and `fd()`, and we can expect that there will be a higher occurence of smaller digits returned by `fd()`. This can partially explain the skew towards $1$ and $2$ that we see in the `ld()` data from the New York Times Article.

# Conclusion

`Benford's Law` is an interesting mathematical property of real-life data sets that has many applications in the real world. `Benford's Law` is an observation that in many real-life sets of numerical data, the leading digit is likely to be small. In sets that obey the law, the number 1 appears as the leading significant digit about 30â€‰% of the time, while 9 usually appears as the leading significant digit less than 5â€‰% of the time. One important application of this property is in detection of accounting fraud, where fabricated figures are more likely to follow a uniform distribuition. By comparing these potentially fabricated figures to behavior expected by `Benford's Law`, it is possible to detect anomalous behavior.

In this project we looked at the occurences of the numbers $1$-$9$ at the first significant digit as well as at the last significant digit for numbers throughout different real-life data sets. While observing these occurences we found that the first significant digit of a number was more likely to be a smaller digit, and that there generally was a uniform distribution of occurences of digits $1$-$9$ for the last significant digit of a number. The number $1$ occured at around a $30$% rate as the first significant digit in each data set, and the number $9$ occured at around $5$% or less in each data set. As these data sets were un-fabricated, the results proved that `Benford's Law` holds true for real-life data sets.
