# Word Frequency Analysis

For analysis of the most frequently occuring words, we will use a dictionary to store data in the format `dictionary = {string:integer}`. The text contains many 'special' characters such as the newline character `\n`, and others like apostrophes, commas, periods, numbers, as well as capitalized words/names. In order to determine a true frequency of each word, we need to remove these special characters in our data analysis. To do this, we will use `regex` to remove any special characters that are inside/directly next to each word. The `regex` pattern will exclude any non-letter from lowercase `a-z` and uppercase `A-Z` using the `^` operator.

# J.Verne - Around the World in Eighty Days

```python
atwDict = {}
pattern = '[^A-Za-z0-9]+'
```

With the `dictionary` set up as well as the `regex` pattern established, we can now read in the text into our notebook for analysis.

```python
with open ('aroundTheWorld.txt','r') as atwFile:
    atwText = atwFile.read()
```

This text had several pages where there are many lines/small paragraphs that were sperated by more than one empty line and therefore had many `\n` characters stacked up next to eachother. To clean this up, every `\n` character was replaced with an empty space (1 character wide), and then a list was by splitting the text string at every space.

```python
atwText = atwText.replace('\n',' ')
```

```python
atwText = atwText.split(' ')
```

With our data cleaned up, it is time to analyze the occurrence of each word. This was simply done using a for loop and our empty `dictionary`. If the word did not exist in the `dictionary`, a new `key-value` pair was inserted into the `dictionary` with the word being the `key`, and the `value` equal to $1$. Here, `re.sub()` was used to find strings that matched our desired pattern. Every word was converted to lowercase using the `str.lower()` method for consistent word analysis.

```python
for word in atwText:
    word = re.sub(pattern, '', word).lower()
    if word in atwDict:
        atwDict[word] += 1
    else:
        atwDict[word] = 1
```

Because we used a `dictionary` to store our data, we had to find a way to sort this data by word occurrence. `Dictionaries` do not have a built in sort method, so to sort this data we had to use list comprehension where we can take advantage of pythons list `sorted()` method.

```python
atwTopWords_list = sorted((value,key) for (key,value) in atwDict.items())[-21:]
```

```python
atwTopWords_list # The most common words in 'Around the World in Eighty Days'
```

It was found that '' is one of the top words, remove this as it is not an actual word. This was cause by replacing every newline character with `' '`.

```python
del(atwTopWords_list[-2])
```

We want to chart these words in the order of frequency rank, reverse the list so highest frequency word is first.

```python
atwTopWords_list.reverse()
```

```python
x_atw = [x[1] for x in atwTopWords_list]
```

```python
y_atw = [y[0] for y in atwTopWords_list]
```

```python
plt.figure(figsize=(14,8))
bar_plot1 = plt.bar(np.arange(len(y_atw)),(y_atw), tick_label=x_atw, width=0.5, color=cm.winter(.1))
plt.xticks(fontsize=15, rotation=45)
plt.ylabel('Word Count')
plt.xlabel('Word')
plt.xticks(rotation=45)
plt.title('Most Frequent Words: Around the World in Eighty Days')
plt.show()
```

# Analysis and Findings

Unsurprisingly, `the` was the most common word used in every text by a large amount. `The` is the most commonly used English word, and studies have found that `the` accounts for about $7$% of all printed English language words. [The](https://en.wikipedia.org/wiki/The)

Looking at the first text, `Around the World in Eighty Days`:

```python
del(atwDict['']) # Remove this 
```

```python
atwWordCount = 0
for x in atwDict:
    atwWordCount += atwDict[x]
    
atwDict['the'] / atwWordCount
```

We see that `Around the World in Eighty Days` is consistent to studies of English-worded texts, with about $7$% of its words being `the`. It was also found that `The Three Musketeers` contain the most `the`'s, appearing over thirteen thousand times in the text.

```python
del(tmDict[''])
```

```python
tmWordCount = 0
for x in tmDict:
    tmWordCount += tmDict[x]
    
tmDict['the'] / tmWordCount
```

```python
del(waldenDict[''])
del(republicDict[''])
```

```python
waldenWordCount = 0
for x in waldenDict:
    waldenWordCount += waldenDict[x]
    
waldenDict['the'] / waldenWordCount
```

```python
republicWordCount = 0
for x in republicDict:
    republicWordCount += republicDict[x]
    
republicDict['the'] / republicWordCount
```

While `The Three Musketeers` had the most frequent usage of the word `the`, it only made up about $5.9$% of the text, lower than the average for most books. `Walden` had `the` appearing $6.4$% of the time while `The Republic` had `the` appearing $5.9$% of the time.

After `the`; `and`, `of` and `to` were some of the top frequently appearing words in each text.

`Around the World in Eighty Days` and `The Republic` both had the same top $4$ most frequent words, `the`, `and`, `of` and `to`.

`Around the World in Eighty Days` and `The Three Musketeers` had the name of a character as one of the top 20 most frequenty appearing words, `Fogg` and `D'Artagnan`.
