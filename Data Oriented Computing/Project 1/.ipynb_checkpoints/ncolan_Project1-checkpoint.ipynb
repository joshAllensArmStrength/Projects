# Introduction

The K-NN algorithm is a supervised machine learning algorithm that is used for both classification and regression of data. It is an algorithm that does not take any assumptions about the distribution of the data that it will classify. This is also called a non-parametric method. K-NN works by taking a data point and considering the 'K' nearest neighbors. It then uses a voting system to determine which class the data point belongs to. The data point is assigned to the class most common among its 'K' nearest neighbors. 

In this report we will design an implementation of the K-NN algorithm and apply it to the **MNIST** database. 
The MNIST database is a large database of handwritten digits that is used to train and test models of machine learning and computer vision. It is composed of 60,000 training images and 10,000 testing images. Each image is of a size of 28x28 pixels, and each image is grayscale. Each image is labeled with the corresponding digit it depicts, ranging from 0 to 9.

This report will explore how making slight modifications to the K-NN algorithm can produce differing levels of accuracy in data classification when trying to classify MNIST images. We will look at how different values of k and different data set sizes will affect the classification accuracy of the k-nn algorithm. We will also look at how well the images in the MNIST dataset are labeled, and whether or not some images are labeled completely incorrect.

# Part 1 - Create a function `knn_classifier()` which implements the K-NN classification algorithm.

Before we can implement our K-NN classification algorithm, we will first need to pull all of the data from the MNIST database and store it in variables `images` and `labels`. Each index of `images` and `labels` corresponds to the same handwritten digit. We will make use of pythons `requests` library to download the necessary files off of the MNIST database, and store them locally in files.

```python
def get_MNIST():
    mnist_url = "http://yann.lecun.com/exdb/mnist/"
    img_file = "ncolan_train-images-idx3-ubyte.gz"
    labels_file = "ncolan_train-labels-idx1-ubyte.gz"

    for fname in [img_file, labels_file]:
        if Path(fname).is_file() :
            print(f"Found: {fname}")
            continue
        print(f"Downloading: {fname}")
        r = requests.get(mnist_url + fname)
        with open(fname, 'wb') as foo:
            foo.write(r.content)
    
    with gzip.open(labels_file, 'rb') as foo:
        e = foo.read()
        
    with gzip.open(img_file, 'rb') as foo:
        f = foo.read()
        
    labels = np.array([b for b in e[8:]])    
    images = np.array([b for b in f[16:]]).reshape(-1, 28*28)
    
    return labels, images 
```

```python
labels, images = get_MNIST()
```

Now that we have our data loaded into 2 variables, `images` and `labels`, we can now create functions necessary to compute the k-nearest neighbors of each point. We will make use of Euclidean distance in order to compute the distance between neighbors. Eucledian distance (or norm-2) is a measure of similarity between two points in a space, calculated as the straight-line distance between them. We can use it to calculate how far away points in the MNIST dataset are from eachother to determine a point's closest neighbors.

The function `dist(X,w)` calculates the Eucledian distance between all of the points of 2 different vectors. The function `n_smallest(a,n)` returns the indices of the `n` smallest values in a vector. In this case we will use `n_smallest(a,n)` to return the shortest `n` distances calculated in `dist(X,w)` in order to find the `k`-closest neighbors of a point in space.

The function `find_kNN(k,data,point)` takes three inputs:
* `k` - the number of nearest neighbors to return 
* `data` - the set of all data points as a vector
* `point` - the index in 'data' of interest 

This function combines these two functions and will return the `k` closest points in `data` to our point of interest, `point`.

```python
def find_kNN(k,data,point):
    def dist(X,w): return np.linalg.norm(X-w, axis=1)
    def n_smallest(a,n): return np.argsort(a)[:n]
    
    if k == -1: 
        return np.array([n_smallest(dist(data[x],data),-1) for x in range(len(data))])
    else:
        return n_smallest(dist(data[point],data),k+1)
```

With the functionality necessary to calculate the k-nearest neighbors created, we can now package it all otgether into one function. The function `knn_classifier(training_data,training_labels,x,k)` takes four inputs:
* `training_data` - The MNIST image data
* `training_labels` - The MNIST labels data
* `x` - The point we are interested in
* `k` - The amount of neighbors to classify

This function returns a tuple of `(label,neighbors)` where:
* `label` - The predicted label of `x`
* `neighbors` - The `k` nearest neighbors of `x`

This function classifies `x` by using the most common occuring neighbor of point `x`. I have added functionality to calculate the distance of all the neighbors of the point by inputting `-1` for `k`.

```python
def knn_classifier(training_data,training_labels,x,k):
    
    def find_kNN(k,data,point):
        def dist(X,w): return np.linalg.norm(X-w, axis=1)
        def n_smallest(a,n): return np.argsort(a)[:n]
        if k == -1: 
            return np.array([n_smallest(dist(data[x],data),-1) for x in range(len(data))])
        else:
            return n_smallest(dist(data[point],data),k+1)
    
    kNeighbors = find_kNN(k,training_data,x)[1:]
    label = stats.mode(training_labels[kNeighbors])[0][0]
    
    return (label, kNeighbors)
```

# Which value of k produces the most accurate prediction?

The k-nearest neighbor classification uses the value of k to determine how many neighbors in the training dataset to use for making predictions. A higher k value means that more neighbors are used for making a prediction, which should result in a prediction that is less likely to be affected by noise in the data. A lower k value means that fewer neighbors are used for making a prediction, which can lead to a very accurate classification, but one that is more suseptible to noise or errors in the data. Therefore, the value of k needs to be carefully chosen to ensure that the best prediction accuracy is achieved. We will look at different values of k and how they affect classification accuracy.

To go about researching which value of k is most accurate, I have written a function `differentK_accuracy(images,labels)`. This function takes 2 inputs:
* `images` - The MNIST image data
* `labels` - The MNIST label data

This will return a list contaning all values of k between 3 and 100 and how accurate the prediction is with that k-value. We are using the `sklearn` library's implementation of the k-NN algorithm, as it is a very efficient and fast method. 

```python
def differentK_accuracy(images,labels):
    
    kVals = np.arange(3,100,1)
    accuracies = []
    
    for k in kVals:
        numCorrect = 0
        knn = KNeighborsClassifier(k)
        knn.fit(images,labels)
        a = knn.predict(images)
        b = (a==labels)
        acc = (b.sum()/len(b)) # np.array().sum() will return the amount of True values in the array 
        accuracies.append((k,acc))
    
    return accuracies
```

# How does the size of the training set affect the classification accuracy?

The size of the training set has a direct impact on the accuracy of KNN classification. The larger the training set, the higher the accuracy of classification should be, as a larger dataset should provide more information for the algorithm to learn from. Conversely, a smaller training set should result in less accurate results. We will use a k value of `3`, as it was found from the previous test that `k = 3` provided the most accurate classification.

For this we will define a new function, `differentTrainingSetSize_accuracy(images,labels,k)` which takes three inputs:
* `images` - The MNIST image data
* `labels` - The MNIST label data
* `k` - The amount of k-neighbors to use

This function will return a list of `[(size,accuracy)]`, which represents the relation of the size of the dataset to the accuracy of the k-nn classifier. This function also uses a `Boolean` mask and `numpy.ndarray.sum()` to find the amount of `True` values, and compares the amount of `True` values to the amount of neighbors to calculate the accuracy.

```python
def differentTrainingSetSize_accuracy(images,labels,k):
    setSize = np.arange(500,len(images),1000) # We will increase the training set by 1000 each time (1.6%)
    accuracies = []
    
    for size in setSize:
        numCorrect = 0
        knn = KNeighborsClassifier(k)
        knn.fit(images[:size],labels[:size])
        a = knn.predict(images[:size])
        b = (a==labels[:size])
        acc = (b.sum()/len(b)) # np.array().sum() will return the amount of True values in the array 
        accuracies.append((size,acc))
        
    return accuracies
```

# What Fraction Of Images Are Classified With Perfect Certainty?

We can see from above that `k = 3` produced the most accurate classification in our data set. However, it is important to realize that when only looking at the nearest 3 neighbors, that errors in the data can have a large effect on the outcome of classification. For that reason, we will use a k value of `5` to see how many images get classified perfectly (the images are classified correctly, and all of their k-nearest neighbors had the same label.)

We will create a function `perfect_accuracy(images,labels,k)`, which takes three inputs:
* `images` - The MNIST image data
* `labels` - The MNIST label data
* `k` - The amount of k-neighbors to use

This function will return a list of the form `[(index,index label,nearest k-neighbors)]` which will show what exact indices of the images were labeled with perfect accuracy, their label, and their k-closest neighbors. This will also use a `Boolean` mask and `numpy.ndarray.sum()`.

```python
def perfect_accuracy(images,labels,k):
    
    perfect = []
    knn = KNeighborsClassifier(k)
    knn.fit(images,labels)
    a = knn.kneighbors(n_neighbors=k)
    b = a[1]
    num = 0
    for i in b:
        c = labels[i] == labels[num]
        if c.sum() == c.size:
            perfect.append((num,labels[num],i))
        num += 1
        
    return perfect 
```

# How Many Images Were Classified With 0% Certainty? i.e. No neighbors had the correct label

We will now look at points in the dataset where all of the points nearest k-neighbors were labeled incorrectly. To do this we will create a function `zero_accuracy(images,labels,k)` whose inputs are:
* `images` - The MNIST image data
* `labels` - The MNIST label data
* `k` - The amount of k-neighbors to use

This function will return a list of `[(index,index label,nearest k-neighbors)]` which are all points whose neighbors have incorrect labels. This will again use a `Boolean` mask and `numpy.ndarray.sum()` and will record points where all of the neighbors had a different label.

```python
def zero_accuracy(images,labels,k):
    
    zero = []
    knn = KNeighborsClassifier(k)
    knn.fit(images,labels)
    a = knn.kneighbors(n_neighbors=k)
    b = a[1]
    num = 0
    for i in b:
        c = labels[i] == labels[num]
        if c.sum() == 0:
            zero.append((num,labels[num],i))
        num += 1
        
    return zero 
```

### Different Training Set Sizes:
#### k = 3:

```python
def differentTrainingSetSize_accuracy(images,labels,k):
    setSize = np.arange(500,len(images),1000) # We will increase the training set by 1000 each time (1.6%)
    accuracies = []
    
    for size in setSize:
        numCorrect = 0
        knn = KNeighborsClassifier(k)
        knn.fit(images[:size],labels[:size])
        a = knn.predict(images[:size])
        b = (a==labels[:size])
        acc = (b.sum()/len(b)) # np.array().sum() will return the amount of True values in the array 
        accuracies.append((size,acc))
        
    return accuracies


b = differentTrainingSetSize_accuracy(images,labels,3)

filename2 = 'ncolan_sklearnKNN_differentSetSizeValues_3KNeighbors.txt'
with open(filename2,'w') as f:
    for i in b:
        f.write('{:} - {:}\n'.format(i[0],i[1]))
        
with open (filename2,'r') as f:
    differentSetSize3_results = np.array([value.strip().split(' - ') for value in f],dtype=float)
    
x_values2 = np.array([i[0] for i in differentSetSize3_results],dtype=int)
y_values2 = np.array([i[1] for i in differentSetSize3_results])*100
x_ticks2 = np.arange(len(x_values2))

plt.figure(figsize=(14,8),dpi=100)
plt.plot(x_ticks2,y_values2,label='Accuracy')
plt.xticks(x_ticks2,x_values2,rotation=60)
plt.xlabel('Size of Training Set')
plt.ylabel('Accuracy of Classification(%)')
plt.title('How Different Training Set Sizes Affect the Accuracy of K-NN Classifier (K = 3)')
plt.legend()
plt.tight_layout()
```

#### k = 5:

```python
c = differentTrainingSetSize_accuracy(images,labels,5)

filename3 = 'ncolan_sklearnKNN_differentSetSizeValues_5KNeighbors.txt'
with open(filename3,'w') as f:
    for i in c:
        f.write('{:} - {:}\n'.format(i[0],i[1]))
        
with open (filename3,'r') as f:
    differentSetSize5_results = np.array([value.strip().split(' - ') for value in f],dtype=float)
    
x_values3 = np.array([i[0] for i in differentSetSize5_results],dtype=int)
y_values3 = np.array([i[1] for i in differentSetSize5_results])*100
x_ticks3 = np.arange(len(x_values3))

plt.figure(figsize=(14,8),dpi=100)
plt.plot(x_ticks3,y_values3,label='Accuracy')
plt.xticks(x_ticks3,x_values3,rotation=60)
plt.xlabel('Size of Training Set')
plt.ylabel('Accuracy of Classification(%)')
plt.title('How Different Training Set Sizes Affect the Accuracy of K-NN Classifier (K = 5)')
plt.legend() 
plt.tight_layout()
```

### Labels Classified With Perfect Certainty:

```python
def perfect_accuracy(images,labels,k):
    
    perfect = []
    knn = KNeighborsClassifier(k)
    knn.fit(images,labels)
    a = knn.kneighbors(n_neighbors=k)
    b = a[1]
    num = 0
    for i in b:
        c = labels[i] == labels[num]
        if c.sum() == c.size:
            perfect.append((num,labels[num],i))
        num += 1
        
    return perfect 

d = perfect_accuracy(images,labels,5)

filename4 = 'ncolan_pointsClassifiedWithPerfectAccuracy.txt'
with open(filename4,'w') as f:
    for i in d:
        f.write('{:}\n'.format(i))
        
with open (filename4,'r') as f:
    perfectAccuracy_results = [value.split('\n') for value in f]
    
len(perfectAccuracy_results)/len(images) * 100
```

### Labels Classified With 0% Certainty:

```python
def zero_accuracy(images,labels,k):
    
    zero = []
    knn = KNeighborsClassifier(k)
    knn.fit(images,labels)
    a = knn.kneighbors(n_neighbors=k)
    b = a[1]
    num = 0
    for i in b:
        c = labels[i] == labels[num]
        if c.sum() == 0:
            zero.append((num,labels[num],i))
        num += 1
        
    return zero 

e = zero_accuracy(images,labels,5)


filename5 = 'ncolan_pointsClassifiedWithZeroAccuracy.txt'
with open(filename5,'w') as f:
    for i in e:
        f.write('{:}\n'.format(i))
        
with open (filename5,'r') as f:
    zeroAccuracy_results = [value.split('\n') for value in f]
    
len(zeroAccuracy_results)
len(zeroAccuracy_results)/len(images)*100
```
